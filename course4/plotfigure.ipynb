```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import xlrd
```

# 1.绘制$y = cos2x$曲线

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.cos(2*x)

fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(x, y, label='cos(2x)')
xticks = [round(i * np.pi, 1) for i in np.arange(-2,3,0.5)]


ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')
ax.set_yticks([-1, 1])
ax.set_xticks(xticks)
ax.set_xticklabels([f'{i}'+ r'$\pi$' if i != 0 else 0 for i in np.arange(-2,3, 0.5)])
ax.axhline(y = 1, c = 'r')
ax.axhline(y = -1, c = 'r')
```

# 2.绘制笛卡尔心形线 $r = 1 + cos(\theta)$

```python
#定义函数
def cardioid(theta):
    return 1 + np.cos(theta)


theta = np.linspace(0, 2 * np.pi, 1000)
r = cardioid(theta)
# 将极坐标转换为笛卡尔坐标
x = r * np.cos(theta)
y = r * np.sin(theta)

# 绘制心形线
fig, ax = plt.subplots(figsize=(6, 6))
plt.plot(x, y)


plt.axis('equal')
plt.title('Cardioid')
```

# 3.使用Matplotlib实现学术成绩分饼图


```python
data = xlrd.open_workbook(r'../data/marks.xls')
table = data.sheets()[0]

matrix = []
for i in range(table.nrows):

    matrix.append(table.row_values(i))

matrix = np.array(matrix)
matrix_df = pd.DataFrame(matrix[1:, 2:], columns = matrix[0, 2:], dtype = np.float64)
mean_score = matrix_df.mean(axis = 0).to_frame().rename(columns={0: 'mean'})
display(mean_score)


fig = plt.figure(figsize=(10, 8))
_ = plt.pie(
    data = mean_score,
    x = 'mean',
    labels = mean_score.index,
    autopct = '%1.1f%%',
    textprops = {'fontsize': 12},
    shadow = True,
    startangle = 90,
    explode = (0, 0.1, 0, 0)   
)
```
